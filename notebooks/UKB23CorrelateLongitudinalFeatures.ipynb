```python
# We copy longitudinal fields to imagenomer for correlation

# Load huge data

import pandas as pd

df = pd.read_csv('/run/media/anton/AntonData1/UKB/Anton_07132023.csv', delimiter=',')

df 
```

```python
from math import isnan

# long_fields = ['f.46', 'f.47',  'f.48', 'f.49', 'f.93', 'f.94', 'f.95', 'f.134', 'f.135']
# long_fields = ['f.136', 'f.137', 'f.864', 'f.874', 'f.884', 'f.894', 'f.904', 'f.914']
# long_fields = ['f.1050', 'f.1060', 'f.1070', 'f.1080', 'f.1090', 'f.1160', 'f.1269', 'f.1279', 'f.1289', 'f.1299', 'f.1309', 'f.1319']
long_fields = ['f.21001', 'f.21002', 'f.23099', 'f.23100', 'f.23101', 'f.23102', 'f.23104', 'f.23105', 'f.23106']

for field in long_fields:
    dct = {}
    for sub in demo['longitudinal']:
        row = df[df['f.eid'] == sub]
        try:
            a = row[f'{field}.2.0'].array[0]
            b = row[f'{field}.3.0'].array[0]
        except:
            continue
        if isnan(a) or isnan(b):
            continue
        diff = b-a
        dct[sub] = diff
    print(len(dct))
    demo[f'{field}.diff3-2'] = dct

print('Complete')
```
